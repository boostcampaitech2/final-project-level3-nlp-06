# 시각화

```python
import numpy as np
import seaborn as sns
from matplotlib import rcParams

sns.set(font="NanumGothic", 
        rc={"axes.unicode_minus":False, 'figure.figsize':(12,4)},
        style='darkgrid')
```

```python
# 회사의 연혁 문서 길이 분포
article_len =  [len(e) for e in dataset['train']['회사의 연혁']]
sns.displot(article_len, height=4, aspect=5)
print(f"min:{np.min(article_len)}", f"max:{np.max(article_len)}", f"mean:{np.mean(article_len):.0f}", f"median:{np.median(article_len):.0f}")
```

```python
# 사업의 개요 문서 길이 분포
article_len =  [len(e) for e in dataset['train']['사업의 개요']]
sns.displot(article_len, height=4, aspect=5)
print(f"min:{np.min(article_len)}", f"max:{np.max(article_len)}", f"mean:{np.mean(article_len):.0f}", f"median:{np.median(article_len):.0f}")
```
