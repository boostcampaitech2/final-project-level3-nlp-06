```python

from datasets import load_dataset
import pandas as pd
import tqdm.notebook as tq

dataset = load_dataset("nlpHakdang/beneficiary", data_files = "dart_v1_3.csv", use_auth_token= "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")
df = pd.DataFrame(dataset['train'])

cols = ['회사의 개요', '회사의 연혁', '사업의 개요', '주요 제품 및 서비스', '주요계약 및 연구개발활동']

clean_col = []

for i in tq.tqdm( range(len(df)) ):
    new_clean_text = ""
    for col in cols:
        new_clean_text
        cont = df.iloc[i][col]
        if cont:
            new_clean_text += "\n" + col + "\n" + cont
    clean_col.append( new_clean_text )
df['clean'] = clean_col
df.head()
```
